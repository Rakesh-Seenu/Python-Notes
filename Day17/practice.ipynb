```python
class User:
    pass
```

```python
user_1 = User() #initialize an object of User class
user_1.first_name = "Harry" #assigning attribute first_name to user_1 object
user_1.last_name = "Potter"

print(user_1.first_name) #output: Harry
print(user_1.last_name)
```

```python
class User1: #defining class User1
    def __init__(self, id,name): #constructor with parameters id and name
        self.id = id #assigning parameter id to instance variable self.id
        self.name = name #assigning parameter name to instance variable self.name
        self.followers = 0 #default value for followers is 0
        self.following = 0 #default value for following is 0

    def follow(self, user):
        user.followers += 1
        self.following += 1
```

```python
user_2 = User1(1,"Hermione Granger") #initialize object user_2 with id and name

print(user_2.id)
print(user_2.name)
```

```python
print(user_2.followers)  # Output: 150
```

```python
user_3 = User1(2, "Ron Weasley")
user_4 = User1(3, "Draco Malfoy")
```

```python
user_3.follow(user_4)  # Ron follows Draco
print(user_3.followers)  # Output: 0
print(user_3.following)  # Output: 1
print(user_4.followers)  # Output: 1
print(user_4.following)  # Output: 0
```
